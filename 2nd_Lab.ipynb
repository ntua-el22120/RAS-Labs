# 🧠 Κατανόηση ενός Convolutional Neural Network (CNN)

Τα Convolutional Neural Networks (CNNs) είναι ο πιο διαδεδομένος τύπος νευρωνικών δικτύων για αναγνώριση εικόνων.  
Χρησιμοποιούνται για αντικείμενα, street signs, πρόσωπα, σκηνές και πολλά άλλα.

Στον οδηγό αυτό θα δούμε:

- Τι κάνει το convolution  
- Πώς λειτουργεί η pooling  
- Πώς συνδέονται όλα στο τέλος σε έναν ταξινομητή  
- Τι συμβαίνει βήμα-βήμα

---

## 🔍 1. Τι είναι ένα CNN;

Ένα CNN παίρνει μια εικόνα (π.χ. 128×128×3) και προσπαθεί να εξάγει χαρακτηριστικά (features).  
Αυτά τα χαρακτηριστικά χρησιμοποιούνται για ταξινόμηση, π.χ. STOP, PRIORITY κ.λπ.

Αποτελείται από:

1. **Convolution Layers** → ανιχνεύουν patterns  
2. **Activation (ReLU)** → κρατάνε τις σημαντικές πληροφορίες  
3. **Pooling Layers** → μειώνουν το μέγεθος  
4. **Fully Connected Layers** → παίρνουν τα features και αποφασίζουν την κατηγορία

---

## 🎨 2. Τι είναι το Convolution

Κάθε convolutional layer εφαρμόζει "φίλτρα" (kernels) πάνω στην εικόνα.  
Κάθε φίλτρο ανιχνεύει διαφορετικό pattern:

- άκρες  
- γωνίες  
- χρωματικές αλλαγές  
- υφές
